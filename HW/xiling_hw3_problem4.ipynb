```python
import torch
from torch import nn
import numpy as np
from pathlib import Path
```

```python
file = Path('book-war-and-peace.txt').read_text('utf-8')
```

```python
def TextOneHot(file, num, seqlen):
  text = ""
  for i in range(1,num):
    seq = file[(i-1)*seqlen:i*seqlen]
    text += seq
  char = set(text)
  m = len(char)
  return m, char
```

```python
m, char = TextOneHot(file, len(file), 25)
print(m)
print(char)
```

```python
int2char = dict(enumerate(char))
char2int = {char: i for i, char in int2char.items()}
```
